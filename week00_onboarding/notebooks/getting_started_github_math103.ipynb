## 2. Git and GitHub

Git is a version control tool. It tracks changes in files.

GitHub is an online platform where Git repositories can be stored, shared, and reviewed.

A repository, often called a repo, is a project folder tracked by Git.

## 3. Recommended Repository Structure

A possible Math 103.1 repository structure is:

```text
math103-text-analytics/
├── README.md
├── notebooks/
│   ├── week01_intro.ipynb
│   ├── week02_vectorization.ipynb
│   └── week03_classification.ipynb
├── data/
│   ├── raw/
│   └── processed/
├── src/
│   └── helper_functions.py
├── outputs/
│   ├── figures/
│   └── tables/
├── reports/
│   ├── week01_reflection.md
│   └── week02_reflection.md
└── genai_logs/
    └── week01_genai_log.md
```

Your actual folder structure may vary, but it should be clear and consistent.

````python
# This Python cell creates a sample folder structure locally.
# You may run it inside a practice folder.

from pathlib import Path

root = Path("math103-text-analytics-demo")

folders = [
    "notebooks",
    "data/raw",
    "data/processed",
    "src",
    "outputs/figures",
    "outputs/tables",
    "reports",
    "genai_logs"
]

for folder in folders:
    (root / folder).mkdir(parents=True, exist_ok=True)

(root / "README.md").write_text("# Math 103.1 Text Analytics Repository\n", encoding="utf-8")

print(f"Created sample folder structure in: {root.resolve()}")
````

## 4. What Should Be Included?

You should usually include:

- Jupyter notebooks;
- Markdown reports;
- Python scripts you wrote;
- small processed datasets, if permitted;
- figures and tables needed for interpretation;
- README files;

## 5. What Should Not Be Included?

Do not upload:

```text

__pycache__/
.env
*.log

```

## 6. The README File

A `README.md` file explains what the repository is about.

Example:

```markdown
# Math 103.1 Text Analytics Repository

This repository contains my weekly notebooks, reports, and experiment logs for Math 103.1: Predictive Analytics for Text.

## Repository Structure

- `notebooks/` contains weekly Jupyter notebooks.

- `reports/` contains weekly written submissions.
- `genai_logs/` contains GenAI use documentation.

## Reproducibility Notes
```

````python
# This cell writes a sample README.md into the demo repository.

from pathlib import Path

root = Path("math103-text-analytics-demo")

readme_text = '''# Math 103.1 Text Analytics Repository

This repository contains my weekly notebooks, reports, and experiment logs for Math 103.1: Predictive Analytics for Text.

## Repository Structure

- `notebooks/` contains weekly Jupyter notebooks.
- `data/` contains raw and processed data when permitted.
- `outputs/` contains figures and tables.
- `reports/` contains weekly written submissions.
- `genai_logs/` contains GenAI use documentation.

## Reproducibility Notes

The notebooks should be run from top to bottom. Random seeds will be documented when used.
'''

(root / "README.md").write_text(readme_text, encoding="utf-8")

print((root / "README.md").read_text(encoding="utf-8"))
````

## 7. Basic GitHub Workflow

A simple workflow is:

```text
Edit files → Save files → Commit changes → Push to GitHub
```

A commit is a saved checkpoint.

A push sends your local commits to GitHub.

## 8. Basic Git Commands

Use these commands in the terminal, not inside a Python cell.

Check status:

```bash
git status
```

Add files:

```bash
git add README.md
```

Add all changed files:

```bash
git add .
```

Commit:

```bash
git commit -m "Add Week 1 notebook and reflection"
```

Push:

```bash
git push
```

Pull latest changes:

```bash
git pull
```

## 9. Suggested Weekly GitHub Routine

For each weekly activity:

1. Create or update the weekly notebook.
2. Add Markdown explanations.
3. Run the notebook from top to bottom.
4. Save the notebook.
5. Write or update the weekly report.
6. Write the GenAI use log, if applicable.
7. Check `git status`.
8. Add the relevant files.
9. Commit with a clear message.
10. Push to GitHub.
11. Open GitHub in a browser and confirm that the files are visible.


## 11. GitHub and Academic Integrity

GitHub makes your work visible and traceable.

You should not:

- copy another student’s notebook and submit it as your own;
- upload private data;
- hide GenAI-generated work;
- erase history to conceal how a submission was made;
- submit code you do not understand.

You may:

- use GitHub to document your progress;
- review your own earlier versions;
- collaborate when collaboration is allowed;
- cite external code or documentation;
- document GenAI support honestly.

## 12. Minimal `.gitignore`

A `.gitignore` file tells Git which files to ignore.

Example:

```text
.DS_Store
__pycache__/
.ipynb_checkpoints/
.env
*.log
```

````python
# This cell writes a sample .gitignore file into the demo repository.


from pathlib import Path

root = Path("math103-text-analytics-demo")

gitignore_text = """# macOS system files
.DS_Store

# Windows system files
Thumbs.db
desktop.ini

# Python cache files
__pycache__/
*.py[cod]

# Jupyter notebook checkpoints
.ipynb_checkpoints/

# Environment and secrets files
.env
.venv/
env/
venv/

# Log files
*.log

# MLflow local tracking directory
mlruns/

# Large or temporary model files
*.joblib
*.pkl
*.pickle
"""

(root / ".gitignore").write_text(gitignore_text, encoding="utf-8")

print((root / ".gitignore").read_text(encoding="utf-8"))
````
